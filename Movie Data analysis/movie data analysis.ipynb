```python
#importing libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns       
import numpy as np
%matplotlib inline
```

```python
#loading data
df = pd.read_csv('tmdb-movies.csv')
df.head()
```

```python
# statistic values for this data 
df.describe()
```

```python
# number of unique values in each column
df.nunique()
```

```python
# number of nulls in each column
df.isnull().sum()
```

DATA CLEANING

```python
# dropping columns with large number of null values 
df.drop(['id','imdb_id', 'homepage','overview'],axis=1,inplace=True)
```

```python
# filling values using mean
df.fillna(df.mean(), inplace = True)
df.info()
```

```python
df.isnull().sum()
```

```python
# to replace all the zero value in coloumn with mean value.
df['popularity']=df['popularity'].replace(0,df['popularity'].mean())
df['revenue']=df['revenue'].replace(0,df['revenue'].mean())
df['runtime']=df['runtime'].replace(0,df['runtime'].mean())
df['budget_adj']=df['budget_adj'].replace(0,df['budget_adj'].mean())
df['revenue_adj']=df['revenue_adj'].replace(0,df['revenue_adj'].mean())
```

```python
# Drop duplicate value
df.drop_duplicates(inplace=True)
```

<i>Do the runtime affect the vote count and popularity?<i>

```python
low = df.query('runtime<=60')
med = df.query('60<runtime<=200')
high = df.query('runtime>200')
```

```python
mean_low_runtime= low['popularity'].mean()
mean_med_runtime = med['popularity'].mean()
mean_high_runtime = high['popularity'].mean()
```

```python
locations = [1,2,3]
heights = [mean_low_runtime , mean_med_runtime ,mean_high_runtime]
labels=['low','med','high']
plt.bar(locations, heights, tick_label = labels)
plt.title('Mean Popularity by all runtime')
plt.xlabel('Runtime')
plt.ylabel('Mean Popularity')
```

```python
mean = df['popularity'].mean()
lowest_popularity =  df.query('popularity < {}'.format(mean))
more_popularity =  df.query('popularity >= {}'.format(mean))
```

```python
df['profit'] = df['revenue'] - df['budget'] # creating a new column for profit
```

```python
locations = [1,2]
heights = [lowest_popularity['profit'].mean() , more_popularity['profit'].mean()]
labels=['low','high']
plt.bar(locations, heights, tick_label = labels)
plt.title('Mean Popularity by all profit')
plt.xlabel('Avg profit')
plt.ylabel('Mean Popularity')
```
