2) Maximum number of patient suffering from which disease? 

```python
num_patient = pd.DataFrame(df[['Hipertension','Diabetes', 'Alcoholism', 'Handcap']].sum())
```

```python
num_patient.reset_index(inplace=True)
```

```python
num_patient.columns=['Diseases', 'number of patients']
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='Diseases',y='number of patients',data=num_patient)
```

```python
df['Gender'].value_counts().plot(kind='pie')
```

```python
df1 = df.groupby('Gender')[['Hipertension','Diabetes', 'Alcoholism', 'Handcap']].sum()
```

```python
df1.reset_index(inplace=True)
```

```python
df1.columns
```

what is the overall appointment show-up vs. no show-up rate?

```python
fig, axs = plt.subplots(ncols=2,figsize=(12,7))
sns.barplot(data=df1[df1['Gender']=='F'], ax=axs[0])
sns.barplot(data=df1[df1['Gender']=='M'], ax=axs[1])
```
